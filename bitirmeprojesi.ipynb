MAKİNE ÖĞRENMESİ

```python
#Veri Setinden Bağımsız ve Bağımlı Değişkenleri Ayırma
X = df_copy.drop("Potability",axis=1)#bağımsız değişkenler
y = df_copy["Potability"]#bağımlı değişken
#Veri Setini Eğitim ve Test Setlerine Ayırma
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2, random_state=7)

```

```python
#standartlaştırma işlemi(normal dağılıma sahip olmasının daha iyi performans göstermesini sağlar.)
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()# öğrenme işlemi sırasında eğitim verilerinin ortalama değerini ve standart sapmasını hesaplama
X_train = ss.fit_transform(X_train)# StandardScaler nesnesini eğitim verilerine (X_train) uygulama
X_test = ss.transform(X_test)# StandardScaler nesnesini test verilerine (X_test) uygulama
```

```python
X_train.shape# X_train'in boyutunu kontrol etme
```

```python
X_test.shape# X_test'in boyutunu kontrol etme
```

```python
from sklearn.neighbors import KNeighborsClassifier #kullanılan makine öğrenmesi modeli
model =KNeighborsClassifier()
model.fit(X_train, y_train) #modelin eğitim verileri üzerinde eğitilmesi
```

```python
from sklearn.metrics import accuracy_score
# Eğitim veri seti üzerinde modelin performansını değerlendirin
train_predictions = model.predict(X_train)
train_accuracy = accuracy_score(y_train, train_predictions)
print(f"Eğitim veri seti doğruluğu: {train_accuracy}")

# Test veri seti üzerinde modelin performansını değerlendirin
test_predictions = model.predict(X_test)
test_accuracy = accuracy_score(y_test, test_predictions)
print(f"Test veri seti doğruluğu: {test_accuracy}")
```

```python
print(test_predictions[0:10])#test_predictions, eğitilmiş modelin test veri seti (X_test) üzerinde yaptığı tahminlerdir.
print(y_test[0:10])# y_test, test veri setinin gerçek (doğru) etiketlerini veya hedef değerlerini içerir.
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV #GridSearchCV kullanarak hiperparametre optimizasyonu ile performans arttırma
param_grid = {'n_neighbors': range(1, 30)}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid.fit(X_train, y_train)
print(f"Best parameters: {grid.best_params_}")
print(f"Best cross-validation score: {grid.best_score_}")
best_model = grid.best_estimator_
test_predictions = best_model.predict(X_test)
print(f"Test accuracy with best model: {accuracy_score(y_test, test_predictions)}")
```
